```python
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.colors import LogNorm
from mpl_toolkits.basemap import Basemap
import numpy as np
from netCDF4 import Dataset
```

```python
# Load data
dataDIR = 'S5P_NRTI_L2__NO2____20200314T210850_20200314T211350_12534_01_010302_20200314T230401.nc'
fh = Dataset(dataDIR, mode='r')
```

```python
fh # show headers
```

```python
fh.groups
```

```python
fh.groups['PRODUCT']
```

```python
print (fh.groups['PRODUCT'].variables.keys()) # variable names!
```

```python
print (fh.groups['PRODUCT'].variables['nitrogendioxide_tropospheric_column_precision'])
```

```python
lons = fh.groups['PRODUCT'].variables['longitude'][:][0,:,:]
lats = fh.groups['PRODUCT'].variables['latitude'][:][0,:,:]
no2 = fh.groups['PRODUCT'].variables['nitrogendioxide_tropospheric_column_precision'][0,:,:]
print (lons.shape)
print (lats.shape)
print (no2.shape)


no2_units = fh.groups['PRODUCT'].variables['nitrogendioxide_tropospheric_column_precision'].units
```

```python
lon_0 = lons.mean()
lat_0 = lats.mean()

plt.figure(figsize=(10,10))
m = Basemap(width=5000000,height=3500000,
            resolution='l',projection='stere',\
            lat_ts=40,lat_0=lat_0,lon_0=lon_0)

xi, yi = m(lons, lats)

# Plot Data
cs = m.pcolor(xi,yi,np.squeeze(no2),norm=LogNorm(), cmap='jet')

# Add Grid Lines
m.drawparallels(np.arange(-80., 81., 10.), labels=[1,0,0,0], fontsize=10)
m.drawmeridians(np.arange(-180., 181., 10.), labels=[0,0,0,1], fontsize=10)

# Add Coastlines, States, and Country Boundaries
m.drawcoastlines()
m.drawstates()
m.drawcountries()

# Add Colorbar
cbar = m.colorbar(cs, location='bottom', pad="10%")
cbar.set_label(no2_units)

# Add Title
plt.title('NO2 in atmosphere')
```
